# **Tokenization**

## **Word Tokenization**
Splits text into words/tokens.

```python
tokens = word_tokenize(text)
```

```python
for w_token in tokens:
    print(w_token)
```

```python
print("Number of Words:", len(tokens))
```

## **Lowercasing**
Converts all words to lowercase.

```python
words_lower = [word.lower() for word in tokens]

print("\nLowercased Words:")
print(words_lower)
```

## **Removing stopwords**
Removes frequent, low-information words.

```python
stop_words = set(stopwords.words('english'))
```

```python
words_no_stopwords = [
    word for word in words_lower
    if word.isalpha() and word not in stop_words]

print("\nAfter Stopword Removal:")
print(words_no_stopwords)
```

## **Stemming**
> Stemming reduces a word to its root/base form by cutting characters from the end.

```python
stemmer = PorterStemmer()
```

```python
stemmed_words = list(
    stemmer.stem(word)
    for word in words_no_stopwords
)

print("\nStemmed Words:")
print(stemmed_words)
```

## **Lemmatization**
> It Converts words to meaningful base form




```python
lemmatizer = WordNetLemmatizer()
```

```python
lemmatized_words = [lemmatizer.lemmatize(word) for word in words_no_stopwords]

print("\nLemmatized Words :")
print(lemmatized_words)
```
